```python
# Import all necessary modules

from google.colab import files
import re
```

```python
fil = files.upload()
```

```python
fil


key = list(fil.keys())
text = str(fil[key[0]])
```

```python
sentence = "Hi, my name is Mike and I like to eat Mike's chocolates"
regex_name = re.compile('Mike')
regex_name.findall(sentence)

names = ['Adam', 'Bob', 'Charles']
for name in names:
  print(regex_name.sub(name,sentence))
```

```python
regex_name.sub('MIKE', sentence)
```

```python
sentence = "Hi, my name is Mike and I like to eat Muke's chocolates"

regex_name = re.compile(r'M\wke')
regex_name.findall(sentence)
```

```python
# Pattern to replace
'\\n\\n1\\n00:05.080 --> 00:09.310\\n'

pattern2replace = r'\\n\\n\d{1,3}\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\n'
regex_timestamps = re.compile(pattern2replace)
regex_timestamps.findall(text)
```

```python
newtext = regex_timestamps.sub(' ', text)
newtext = newtext[9:-1]
```

```python
# Replace 4 letter words

words = newtext.split()

for wordi in range(len(words)):
  if len(words[wordi]) == 4 and not words[wordi][-1] == '.':
    words[wordi] = words[wordi][0] + '***'
cleantext = ' '.join(words)
```

```python
# Write to object and download

with open('clean_transcript.txt','w') as f:
  f.write(cleantext)
files.download('clean_transcript.txt')
```

```python
# Bonus: Scrambling words

text = "I have often wondered wheter it's possible to read an entire book faster if the useless letters were removed"
words = text.split()
```

```python
# Test scrambling

import numpy as np

w = 'removed'
idx = np.random.permutation(len(w))
''.join([w[i] for i in idx])
```

```python
# Completely scramble all letters

for i in range(len(words)):
  thisword = words[i]
  idx = np.random.permutation(len(thisword))
  words[i] =''.join([thisword[i] for i in idx])
' '.join(words)
   
```

```python
# Scramble all except first and last letters

words = text.split()

for i in range(len(words)):
  # ignore short words
  if len(words[i]) < 4:
    continue
  thisword = words[i][1:-1]
  idx = np.random.permutation(len(thisword))
  words[i] = words[i][0]+''.join([thisword[i] for i in idx]) \
             +words[i][-1]
' '.join(words)
   
```

```python
# Random scrambling
import string
allletters = string.ascii_lowercase

words = text.split()

for i in range(len(words)):
  # ignore short words
  if len(words[i]) < 4:
    continue
  thisword = words[i][1:-1]
  idx = np.random.randint(0,len(allletters),len(thisword))
  words[i] = words[i][0]+''.join([allletters[i] for i in idx]) \
             +words[i][-1]
' '.join(words)
```
