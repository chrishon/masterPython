```python
import  glob
import re
```

```python
pyfiles = glob.glob('*ipynb')
pyfiles
```

```python
# import each file
allfiles = []

for fil in pyfiles:
  fid = open(fil,'r')
  content = fid.read()
  allfiles.append(content)
  #allfiles.append(open(fil,'r').read()) # oneline
allfiles[0]
```

```python
s = 'My name is Mike'
# find words that end with letter e

re.findall(r'\w*e\b',s)

# find words that start with M

re.findall(r'\bM\w*',s)
```

```python
t = '"alist = list(range(5,11)\\n"'
re.findall(r'\w*\( ?\b',t)
```

```python
fi = 0
allfuncalls = []
for fi,thisfile in enumerate(allfiles):
  funwords = re.findall(r'\w*\( ?\b',thisfile)
  funwords = [i.strip()[:-1] for i in funwords]
  funwords = [i for i in funwords if len(i)>0]


  for f in funwords:
    allfuncalls.append([f,pyfiles[fi][:-6]])
  fi += 1
allfuncalls
```

```python
uniquenames = []
for i in allfuncalls:
  if i[0] not in uniquenames:
    uniquenames.append(i[0])
```

```python
indexdict = {}

for i in allfuncalls:
  # create a new key if it doesn't exist
  if i[0] not in indexdict:
    indexdict[i[0]] = []

    # add filename to values list
    if i[1] not in indexdict[i[0]]:
      indexdict[i[0]].append(i[1])
```

```python
sorted(indexdict.keys())
```

```python
fid = open('function_index.txt','w')
for key in sorted(indexdict):
  # function call in first column
  fid.write(f'{key};')
  # files in second cell
  tmps = str(indexdict[key]).replace('[','').replace(']','').replace("'",'')
  fid.write(f'{tmps}')

  #newline
  fid.write('\n')
fid.close()
```
