```python
def mymean(data):
  return  sum(data)/len(data)

def mymedian(data):
  data.sort()
  N = len(data)
  if N%2==1:
    mymedian = data[N//2]
  else:
    mymedian = (data[N//2-1]+data[N//2+1])/2

  return mymedian

def mymode(data):
  uniquevals = list(set(data))
  valcounts = [0]*len(uniquevals)
  for d in data:
    idx = [i for i,u in enumerate(uniquevals) if d==u][0]
    valcounts[idx] +=1 
  table = {}

  for i in range(len(uniquevals)):
    table[(uniquevals[i])]  = valcounts[i]
  maxcount = max(list(table.values()))

  mode = [k for k,v in table.items() if v==maxcount]
  return mode 


def mystd(data):
  meanval = sum(data)/len(data)
  summation = sum([(d-meanval)**2 for d in data])
  std = (summation/(N-1))**(1/2)

  return std
```

```python
N = 20
data = np.random.randint(10,41,size = N)

fid = open('stats.csv','w')
fid.write('mean;' + str(mymean(data))+'\n')
fid.write('median;' + str(mymedian(data))+'\n')
fid.write('mode;' + str(mymode(data))+'\n')
fid.write('std;' + str(mystd(data))+'\n')

fid.close()
```
